### ЗАДАНИЕ: воссоздайте график Scatter Plot, показанный ниже.

**График scatterplot пытается показать зависимость между количеством дней на работе (days employed) и возрастом человека (DAYS_BIRTH) для всех, кроме безработных. Обратите внимание, что для этого графика Вам нужно будет сначала удалить безработных людей из набора данных. <img src="task_one.jpg">


```python
# НАПИШИТЕ ВАШ КОД В ЭТОЙ ЯЧЕЙКЕ
data = df
data['DAYS_EMPLOYED'] = abs(df['DAYS_EMPLOYED'])
data['DAYS_BIRTH'] = abs(data['DAYS_BIRTH'])
data = data[data['DAYS_EMPLOYED']!=0]

sns.scatterplot(x='DAYS_BIRTH', y='DAYS_EMPLOYED', data=data, alpha=0.01)
plt.ylim(0,18000)
```

### Задание: воссоздайте приведённый ниже график Distribution Plot:

<img src="DistPlot_solution.png">

**Обратите внимание, что Вы должны будете найти способ вычислить признак "Age in Years" на основе одной из имеющихся колонок. Подумайте об этом внимательно. По поводу стилизации графика не волнуйтесь.**

```python
# НАПИШИТЕ ВАШ КОД В ЭТОЙ ЯЧЕЙКЕ
from datetime import datetime, timedelta

current_date = datetime.now()
data['AGE_IN_YEARS'] = current_date.year - (current_date - pd.to_timedelta(data['DAYS_BIRTH'], unit='D')).dt.year 
graph = sns.displot(x='AGE_IN_YEARS', data=data, color='red')
graph.set_axis_labels('Age in Years')
```

### ЗАДАНИЕ: воссоздайте приведённый ниже график Categorical:

<img src='catplot_solution.png'>

**Этот график отображает информацию только для *половины всех людей*, имеющих доход в этом наборе данных. Причём мы берём "нижнюю" половину - тех, кто зарабатывает меньше. Графики boxplot отображены для каждого значения колонки NAME_FAMILY_STATUS, показывая распределения их общего дохода. Раскраска hue сделана по колонке "FLAG_OWN_REALTY". Замечание: чтобы взять только нижнюю половину распределения дохода, Вам нужно будет преобразовать dataframe или взять только его часть *перед* тем, как рисовать график.**

```python
# НАПИШИТЕ ВАШ КОД В ЭТОЙ ЯЧЕЙКЕ
data = data.nsmallest(n=int(0.5*len(data)), columns='AMT_INCOME_TOTAL', keep='first')
sns.boxplot(y='AMT_INCOME_TOTAL', x='NAME_FAMILY_STATUS', data=data, hue='FLAG_OWN_REALTY')
plt.legend(loc=(1.05,0.5))
```

### ЗАДАНИЕ: воссоздайте приведённый ниже график Heat Map:

<img src='heatmap_solution.png'>

**Эта тепловая карта heatmap показывает корреляцию между колонками в наборе данных. Корреляцию можно посчитать с помощью метода .corr().  Обратите внимание, что колонка FLAG_MOBIL имеет NaN-корреляцию со всеми другими колонками - это видно в результатах метода .corr(). Поэтому Вам следует удалить колонку FLAG_MOBIL перед построением графика.**

```python
data_new = df
data_new = data_new.drop(['FLAG_MOBIL', 'CODE_GENDER'],axis=1)
data_new = data_new[['ID','CNT_CHILDREN','AMT_INCOME_TOTAL','DAYS_BIRTH','DAYS_EMPLOYED','FLAG_WORK_PHONE','FLAG_PHONE','FLAG_EMAIL','CNT_FAM_MEMBERS']]

data_new = data_new.corr()
```

```python
# НАПИШИТЕ ВАШ КОД В ЭТОЙ ЯЧЕЙКЕ
sns.heatmap(data=data_new)
```
